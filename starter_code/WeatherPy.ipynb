```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
print(url)
#(json.dumps(response, sort_keys=True, indent= 4))
```

## Perform API Calls

```python


#declaring variables
city_name=[]
city_temp =[]
city_latt =[]
city_humidity=[]
city_cloudiness=[]
wind_speed=[]

counter = 1
print(" Begining Data Retrieval")
print("--------------------------")
for city in cities:
   # print(f"Processing Record #{counter} of Set {counter} | {city_name}")
    try:
        response = requests.get(f"{url}&q={city}").json()
        city_name.append(response["name"])
        city_temp.append(response["main"]["temp"])
        city_latt.append(response["coord"]["lat"])
        city_humidity.append(response["main"]["humidity"])
        city_cloudiness.append(response["clouds"]["all"])
        wind_speed.append(response["wind"]["speed"])
#print(json.dumps(weather_response, sort_keys=True, indent= 4))
        
        print(f"Processing Record #{counter} of Set {counter} | {city}")
        print(f"Weather Url: {url}&q={city}")
        counter = counter +1
        
    except:
        print(f"Ooops! city not found!!...skipping")
    

```

```python
#creating a pandas df with the derived entities
cities_df = pd.DataFrame({
    "City": city_name,
    "City Temperature": city_temp,
    "City Latitude": city_latt,
    "City Cloudiness":city_cloudiness,
    "City Humidity": city_humidity,
    "City Wind": wind_speed
})
cities_df.head(10)
```

```python

plt.scatter(cities_df["City Latitude"],cities_df["City Temperature"], marker="o",color="skyblue", 
            edgecolors="black", alpha=0.70,) #label="Rural", linewidth = 1.5 s=cities_df["City"]*10)
plt.xlabel(r'Latitude', fontsize=10)
plt.ylabel(r'Max Temperature(F)', fontsize=10)
plt.title('City Latitude vs. Max Temperature (09/26/18)')
plt.ylim(0,120)

plt.grid(True)
plt.savefig("../Images/lati_vs_temp.png")
```
